I found out Morin chapter 9 problem 37 is the same as the 2018 final exam problem 4. Though Morin did not give the solution, he says you can check by showing theta = 61.9 degree gives the largest angle difference. Unlike our final, that problem expects us to use the actual formula for moment of inertia of the sphere. Other than that, everything is the same, and I showed my answer is correct.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
m = 2
R = 4
I = 2/5*m*R**2
```

```python
theta = np.arange(0,np.pi/2,0.01)
```

```python
phi = np.arccos((I + m*(R**2)*(np.cos(theta))**2)/np.sqrt(I**2 + m**2 * R**4 * np.cos(theta)**2 + 2*I*m*(R**2) * np.cos(theta)**2))
```

```python
plt.plot(theta,phi)
plt.axvline(x = 61.9*(np.pi/180))
```
